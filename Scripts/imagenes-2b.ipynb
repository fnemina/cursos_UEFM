```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Firmas primer ejercicio
l = np.array([490,560,665,842])
veg = np.array([0.01,0.04,0.02,0.33])
h2o = np.array([0.03,0.02,0.007,0.003])

plt.plot(l,h2o,'oC0',label="Agua")
plt.plot(l,veg,'oC2',label="Vegetaci√≥n")
plt.legend()
plt.xlabel("Longitud de onda [nm]")
plt.ylabel("Reflectancia")
plt.savefig("p1.png",dpi=300)
plt.show()
```
